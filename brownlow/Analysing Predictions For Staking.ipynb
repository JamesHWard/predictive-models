```python
import pandas as pd
pd.set_option('display.max_rows', 2000)
```

```python
preds = pd.read_csv('data/brownlow_predictions_2018_round_by_round.csv')
```

```python
agg_preds = (preds.groupby(['player', 'team'], as_index=False)
 .predicted_votes_scaled
 .sum()
 .sort_values(by='predicted_votes_scaled', ascending=False)
 .reset_index(drop=True))
```

```python
agg_preds.head(10)
```

```python
team_votes = (preds.groupby('team', as_index=False)
 .predicted_votes_scaled
 .sum()
 .sort_values(by='predicted_votes_scaled', ascending=False)
 .reset_index(drop=True))
```
